```python
# 1.2 Data Cleansing - Sreynich

# ***Check if there's missing values in all columns***
nan_count = myData.isna().sum()
print(nan_count)

#In the output, We can see that the dataset have no missing values.
```

```python
# ***Encoding categorical data values (Transforming object data types to integers)***

#Encode 'Sleep Disorder' column: 1 for 'Sleep Apnea' or 'Insomnia', and 0 for 'None' 
myData['Sleep Disorder Binary Classification'] = myData['Sleep Disorder'].map(lambda x: 1 if x in ['Sleep Apnea', 'Insomnia'] else 0)

#Encode 'Gender' column: 1 for 'Female', and 0 for 'Male'
myData['Gender Numeric'] = myData['Gender'].map(lambda x: 1 if x in ['Female'] else 0)
# labelencoder = LabelEncoder()
# myData.iloc[:,1]= labelencoder.fit_transform(myData.iloc[:,1].values)

#Encode 'BMI Category' column:
bmi_mapping = {
    'Normal': 0,
    'Normal Weight': 0,
    'Overweight': 1,
    'Obese': 1,
}
myData['BMI Category Numeric'] = myData['BMI Category'].map(bmi_mapping)

#Encode 'Occupation Numeric' column:
le = LabelEncoder()
myData['Occupation Numeric'] = le.fit_transform(myData['Occupation'])
```

```python
# ***Check the unique values of Sleep Disorder***
myData["Sleep Disorder Binary Classification"].unique()
```

```python
# ***Get a count of the number of sleep disorder***
myData['Sleep Disorder Binary Classification'].value_counts()
#In this dataset, People who did not have sleep disorder(219) is slightly greater than did(155).
```

```python
# 3.Data Visualization - Seavhong

# Countplot for Gender distribution
sns.countplot(x='Gender Numeric', data=myData)
plt.title('Gender Distribution')
plt.show()

# Histogram for Age distribution
plt.figure(figsize=(8, 6))
sns.histplot(myData['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()

# Boxplot for Sleep Duration by Gender
plt.figure(figsize=(8, 6))
sns.boxplot(x='Gender Numeric', y='Sleep Duration', data=myData)
plt.title('Sleep Duration by Gender')
plt.xlabel('Gender')
plt.ylabel('Sleep Duration')
plt.show()

# Pairplot to explore relationships between numerical columns
sns.pairplot(myData[['Age', 'Sleep Duration', 'Physical Activity Level', 'Stress Level']])
plt.suptitle('Pairplot of Numerical Columns')
plt.show()

# Countplot for Sleep Disorder
plt.figure(figsize=(8, 6))
sns.countplot(x='Sleep Disorder Binary Classification', data=myData)
plt.title('Sleep Disorder Distribution')
plt.xlabel('Sleep Disorder Binary Classification')
plt.ylabel('Count')
plt.show()

# Countplot Matrix of Gender and Occupation by BMI Category
sns.catplot(x='Gender Numeric', hue='Occupation', kind='count', data=myData)
plt.suptitle('Countplot Matrix of Gender and Occupation by BMI Category', y=1.02)
plt.show()
```

```python
# Drop the 'Person ID' column

df = myData.drop(columns=['Person ID','Occupation','BMI Category','Blood Pressure','Sleep Disorder','Gender'])

# Correlation Heatmap (excluding non-numeric columns)
plt.figure(figsize=(10, 8))
numeric_columns = df.select_dtypes(include=['number']).columns
correlation_matrix = df[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)
plt.title('Correlation Heatmap')
plt.show()
```
